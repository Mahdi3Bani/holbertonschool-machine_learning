```python
import sys

"""takes in input from the user with
the prompt Q: and prints A: """
while True:
    try:
        sentence = input("Q: ")
        sentence = sentence.lower()
        if sentence in ['exit', 'quit', 'goodbye', 'bye']:
            print("A: Goodbye")
            sys.exit()
        else:
            print("A: ")
    except (KeyboardInterrupt, EOFError):
        print("A: Goodbye")
        sys.exit()
```
